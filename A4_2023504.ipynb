## 1. Dataset Selection & Description

### Dataset 1: Product Categories (Tree Visualization)

**Source:** Synthetic e-commerce hierarchy  
**Description:** A hierarchical product catalog representing a typical e-commerce platform structure.

**Nodes Represent:**
- Each node represents a product category or subcategory
- Root node: "Electronics" (top-level category)
- Internal nodes: Category groups (e.g., "Computers", "Mobile Devices")
- Leaf nodes: Specific product types (e.g., "Laptops", "Smartphones")

**Edges Represent:**
- Parent-child relationships (hierarchy)
- An edge from parent category to subcategory indicates containment
- Direction flows from general to specific categories

**Why suitable for Tree Visualization:**
- Natural hierarchical structure
- Clear parent-child relationships
- Multiple levels of hierarchy
- Ideal for testing different tree layouts

---

### Dataset 2: Flight Routes Network (Graph Visualization)

**Source:** OpenFlights.org / NetworkX built-in datasets  
**Description:** A transportation network representing flight routes between major airports globally.

**Nodes Represent:**
- Each node represents an airport
- Identified by airport code (e.g., "JFK", "LAX", "LHR")
- Node attributes: airport name, city, country

**Edges Represent:**
- Flight routes connecting airports
- An edge between two airports indicates direct flight routes
- Edge weights can represent distance or number of flights

**Why suitable for Graph Visualization:**
- Complex network with multiple interconnections
- Non-hierarchical structure (not a tree)
- Multiple paths between nodes (cycles)
- Ideal for layout algorithms (force-directed, spectral, etc.)

```python
import pandas as pd
import numpy as np
import networkx as nx
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported successfully!")
```

### 2.1 Create Product Categories Tree Dataset

```python
# Create product categories tree as edges (parent, child) pairs
tree_edges = [
    # Electronics root
    ('Electronics', 'Computers'),
    ('Electronics', 'Mobile Devices'),
    ('Electronics', 'Accessories'),
    ('Electronics', 'Audio'),
    ('Electronics', 'Cameras'),
    
    # Computers subcategories
    ('Computers', 'Laptops'),
    ('Computers', 'Desktops'),
    ('Computers', 'Tablets'),
    ('Computers', 'Components'),
    
    # Mobile Devices subcategories
    ('Mobile Devices', 'Smartphones'),
    ('Mobile Devices', 'Phone Accessories'),
    ('Mobile Devices', 'Smartwatches'),
    
    # Accessories subcategories
    ('Accessories', 'Chargers'),
    ('Accessories', 'Cables'),
    ('Accessories', 'Cases'),
    
    # Audio subcategories
    ('Audio', 'Headphones'),
    ('Audio', 'Speakers'),
    ('Audio', 'Microphones'),
    
    # Cameras subcategories
    ('Cameras', 'DSLR'),
    ('Cameras', 'Mirrorless'),
    ('Cameras', 'Action Cameras'),
]

# Create NetworkX tree from edges
tree = nx.DiGraph()
tree.add_edges_from(tree_edges)

print(f"Tree Dataset Created:")
print(f"  - Number of nodes: {tree.number_of_nodes()}")
print(f"  - Number of edges: {tree.number_of_edges()}")
print(f"  - Root node: Electronics")
print(f"\nTree Structure (first few levels):")
print(f"  {dict(tree.out_degree())}")
```

### 2.2 Create Flight Routes Graph Dataset

```python
# Create flight routes graph (realistic major airports)
flight_edges = [
    # US Hubs
    ('JFK', 'LAX'), ('JFK', 'ORD'), ('JFK', 'MIA'), ('JFK', 'ATL'),
    ('LAX', 'ORD'), ('LAX', 'DFW'), ('LAX', 'SFO'),
    ('ORD', 'ATL'), ('ORD', 'DFW'), ('ORD', 'MIA'),
    ('ATL', 'MIA'), ('ATL', 'DFW'), ('ATL', 'DEN'),
    ('DFW', 'DEN'), ('DFW', 'SFO'), ('DFW', 'SEA'),
    
    # International connections
    ('JFK', 'LHR'), ('JFK', 'CDG'), ('JFK', 'FCO'),
    ('LAX', 'NRT'), ('LAX', 'HND'), ('LAX', 'PEK'),
    ('ORD', 'LHR'), ('ORD', 'CDG'), ('ORD', 'DUB'),
    ('MIA', 'CDG'), ('MIA', 'BOG'), ('MIA', 'SJO'),
    ('SFO', 'NRT'), ('SFO', 'HND'), ('SFO', 'SYD'),
    
    # European connections
    ('LHR', 'CDG'), ('LHR', 'FCO'), ('LHR', 'DUB'), ('LHR', 'AMS'),
    ('CDG', 'FCO'), ('CDG', 'AMS'), ('CDG', 'ZRH'),
    ('FCO', 'AMS'), ('FCO', 'ZRH'),
    
    # Asian connections
    ('NRT', 'HND'), ('NRT', 'PEK'), ('NRT', 'ICN'), ('NRT', 'SYD'),
    ('HND', 'PEK'), ('HND', 'ICN'), ('HND', 'SYD'), ('HND', 'SIN'),
    ('PEK', 'ICN'), ('PEK', 'SIN'), ('PEK', 'BKK'),
    ('ICN', 'SIN'), ('ICN', 'BKK'),
    
    # Make edges bidirectional (flights go both ways)
]

# Create undirected graph (flights can go both directions)
flight_graph = nx.Graph()
flight_edges_bidirectional = flight_edges + [(b, a) for a, b in flight_edges]
flight_graph.add_edges_from(flight_edges_bidirectional)

print(f"Flight Routes Dataset Created:")
print(f"  - Number of airports (nodes): {flight_graph.number_of_nodes()}")
print(f"  - Number of routes (edges): {flight_graph.number_of_edges()}")
print(f"  - Number of unique airports: {flight_graph.number_of_nodes()}")
print(f"\nAirports included:")
print(sorted(flight_graph.nodes()))
print(f"\nNetwork density: {nx.density(flight_graph):.3f}")
print(f"Average degree: {sum(dict(flight_graph.degree()).values()) / flight_graph.number_of_nodes():.2f}")
```

## 3. Data Preparation & Preprocessing

### 3.1 Tree Dataset Preparation

```python
# Analyze tree structure for visualization
print("TREE DATASET PREPARATION")
print("=" * 70)

# Get tree information
tree_nodes = list(tree.nodes())
tree_edges_list = list(tree.edges())

print(f"\n1. Tree Structure Analysis:")
print(f"   - Total nodes: {len(tree_nodes)}")
print(f"   - Total edges: {len(tree_edges_list)}")
print(f"   - Root node: Electronics")

# Calculate depth information
root = 'Electronics'
depths = nx.shortest_path_length(tree, source=root)
depth_dict = dict(depths)

print(f"\n2. Depth Information:")
for node, depth in sorted(depth_dict.items(), key=lambda x: x[1]):
    print(f"   - {node}: depth {depth}")

# Create position data for hierarchy layout
print(f"\n3. Hierarchy Levels:")
levels = {}
for node, depth in depth_dict.items():
    if depth not in levels:
        levels[depth] = []
    levels[depth].append(node)

for level in sorted(levels.keys()):
    print(f"   - Level {level}: {levels[level]}")

# Get children for each node
print(f"\n4. Parent-Child Relationships:")
for node in tree.nodes():
    children = list(tree.successors(node))
    if children:
        print(f"   - {node} → {children}")

print(f"\n✓ Tree dataset is ready for visualization!")
```

### 3.2 Graph Dataset Preparation

```python
# Analyze flight network structure
print("\nGRAPH DATASET PREPARATION")
print("=" * 70)

# Get graph information
graph_nodes = list(flight_graph.nodes())
graph_edges_list = list(flight_graph.edges())

print(f"\n1. Graph Structure Analysis:")
print(f"   - Total airports (nodes): {len(graph_nodes)}")
print(f"   - Total routes (edges): {len(graph_edges_list)}")
print(f"   - Network connectivity: {'Connected' if nx.is_connected(flight_graph) else 'Disconnected'}")

# Degree analysis
degrees = dict(flight_graph.degree())
print(f"\n2. Degree Analysis (Hub Analysis):")
sorted_degrees = sorted(degrees.items(), key=lambda x: x[1], reverse=True)
for airport, degree in sorted_degrees[:10]:  # Top 10 hubs
    print(f"   - {airport}: {degree} routes")

# Shortest paths and distances
print(f"\n3. Network Distances:")
if nx.is_connected(flight_graph):
    diameter = nx.diameter(flight_graph)
    avg_shortest_path = nx.average_shortest_path_length(flight_graph)
    print(f"   - Network diameter: {diameter}")
    print(f"   - Average shortest path: {avg_shortest_path:.2f}")
else:
    # Get largest component info
    largest_cc = max(nx.connected_components(flight_graph), key=len)
    subgraph = flight_graph.subgraph(largest_cc)
    print(f"   - Largest connected component: {len(subgraph)} nodes")
    print(f"   - Network diameter (largest component): {nx.diameter(subgraph)}")

# Betweenness centrality (important hubs)
print(f"\n4. Airport Importance (Betweenness Centrality):")
betweenness = nx.betweenness_centrality(flight_graph)
sorted_betweenness = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)
for airport, importance in sorted_betweenness[:5]:
    print(f"   - {airport}: {importance:.3f}")

# Create node and edge lists for visualization
print(f"\n5. Data Structure for Visualization:")
print(f"   - Nodes as DataFrame:")
node_list = pd.DataFrame({
    'airport_code': list(graph_nodes),
    'degree': [degrees[node] for node in graph_nodes],
    'betweenness': [betweenness[node] for node in graph_nodes]
})
print(f"     Shape: {node_list.shape}")

print(f"\n   - Edges as DataFrame:")
edge_list = pd.DataFrame(list(graph_edges_list), columns=['source', 'target'])
print(f"     Shape: {edge_list.shape}")

print(f"\n✓ Graph dataset is ready for visualization!")
```

### 3.3 Pre-compute Layout Positions for Performance

```python
# Pre-compute layout positions for both datasets
print("\nPRE-COMPUTING LAYOUT POSITIONS")
print("=" * 70)

# ---- TREE LAYOUTS ----
print("\n1. Tree Dataset Layouts:")

# Layout 1: Hierarchical (Top-down layout)
print("   - Computing Hierarchical Layout...")
tree_hierarchical_pos = {}
root = 'Electronics'
levels = {}
depths = nx.shortest_path_length(tree, source=root)

for node, depth in dict(depths).items():
    if depth not in levels:
        levels[depth] = []
    levels[depth].append(node)

# Position nodes based on level and order
for level, nodes in levels.items():
    for i, node in enumerate(nodes):
        x = (i - len(nodes) / 2) * 2
        y = -level * 3
        tree_hierarchical_pos[node] = (x, y)

print(f"     ✓ {len(tree_hierarchical_pos)} nodes positioned")

# Layout 2: Spring layout (force-directed)
print("   - Computing Spring Layout...")
tree_spring_pos = nx.spring_layout(tree, k=2, iterations=50, seed=42)
print(f"     ✓ {len(tree_spring_pos)} nodes positioned")

# ---- GRAPH LAYOUTS ----
print("\n2. Graph Dataset Layouts:")

# Layout 1: Force-directed (spring)
print("   - Computing Force-Directed Layout...")
graph_spring_pos = nx.spring_layout(flight_graph, k=2, iterations=50, seed=42)
print(f"     ✓ {len(graph_spring_pos)} nodes positioned")

# Layout 2: Circular layout
print("   - Computing Circular Layout...")
graph_circular_pos = nx.circular_layout(flight_graph)
print(f"     ✓ {len(graph_circular_pos)} nodes positioned")

# Store all positions in a dictionary for easy access
layout_positions = {
    'tree': {
        'hierarchical': tree_hierarchical_pos,
        'spring': tree_spring_pos
    },
    'graph': {
        'spring': graph_spring_pos,
        'circular': graph_circular_pos
    }
}

print(f"\n✓ All layout positions pre-computed and stored!")
print(f"\nAvailable layouts:")
print(f"  - Tree: Hierarchical, Spring")
print(f"  - Graph: Spring (Force-Directed), Circular")
```

### 3.4 Summary: Data Ready for Visualization

```python
print("\n" + "=" * 70)
print("DATA PREPARATION COMPLETE!")
print("=" * 70)

print("\n✓ TREE DATASET (Product Categories)")
print(f"  - Nodes: {tree.number_of_nodes()}")
print(f"  - Edges: {tree.number_of_edges()}")
print(f"  - Layouts ready: Hierarchical, Spring")
print(f"  - Data structure: NetworkX DiGraph")

print("\n✓ GRAPH DATASET (Flight Routes)")
print(f"  - Nodes: {flight_graph.number_of_nodes()}")
print(f"  - Edges: {flight_graph.number_of_edges()}")
print(f"  - Layouts ready: Spring (Force-Directed), Circular")
print(f"  - Data structure: NetworkX Graph (undirected)")

print("\n✓ All data structures and layouts are prepared!")
print("  Ready to proceed to Step 3: Visualization\n")
```

## 4. Tree Visualization

### 4.1 Tree Visualization Layout 1: Hierarchical Layout

```python
# Tree Hierarchical Layout Visualization
print("Creating Tree Visualization 1: Hierarchical Layout...")

# Get positions
pos_hier = layout_positions['tree']['hierarchical']

# Prepare edge traces for hierarchical layout
edge_x_hier = []
edge_y_hier = []

for edge in tree.edges():
    x0, y0 = pos_hier[edge[0]]
    x1, y1 = pos_hier[edge[1]]
    edge_x_hier.extend([x0, x1, None])
    edge_y_hier.extend([y0, y1, None])

# Create edge trace
edge_trace_hier = go.Scatter(
    x=edge_x_hier, y=edge_y_hier,
    mode='lines',
    line=dict(width=2, color='#888'),
    hoverinfo='none',
    showlegend=False
)

# Prepare node trace
node_x_hier = []
node_y_hier = []
node_text_hier = []
node_color_hier = []

depths = dict(nx.shortest_path_length(tree, source='Electronics'))

for node in tree.nodes():
    x, y = pos_hier[node]
    node_x_hier.append(x)
    node_y_hier.append(y)
    node_text_hier.append(node)
    # Color by depth
    node_color_hier.append(depths[node])

# Create node trace
node_trace_hier = go.Scatter(
    x=node_x_hier, y=node_y_hier,
    mode='markers+text',
    text=node_text_hier,
    textposition='top center',
    hoverinfo='text',
    hovertext=node_text_hier,
    marker=dict(
        size=20,
        color=node_color_hier,
        colorscale='Viridis',
        showscale=True,
        colorbar=dict(title="Depth Level"),
        line_width=2
    )
)

# Create figure
fig_tree_hier = go.Figure(data=[edge_trace_hier, node_trace_hier])

fig_tree_hier.update_layout(
    title='<b>Tree Visualization: Hierarchical Layout</b><br><sub>Product Categories Hierarchy</sub>',
    showlegend=False,
    hovermode='closest',
    margin=dict(b=20, l=5, r=5, t=40),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    height=700,
    width=1000,
    plot_bgcolor='rgba(240, 240, 240, 0.5)'
)

fig_tree_hier.show()
print("✓ Tree Hierarchical Layout visualization created!")
```

### 4.2 Tree Visualization Layout 2: Spring (Force-Directed) Layout

```python
# Tree Spring (Force-Directed) Layout Visualization
print("Creating Tree Visualization 2: Spring Layout...")

# Get positions
pos_spring = layout_positions['tree']['spring']

# Prepare edge traces
edge_x_spring = []
edge_y_spring = []

for edge in tree.edges():
    x0, y0 = pos_spring[edge[0]]
    x1, y1 = pos_spring[edge[1]]
    edge_x_spring.extend([x0, x1, None])
    edge_y_spring.extend([y0, y1, None])

# Create edge trace
edge_trace_spring = go.Scatter(
    x=edge_x_spring, y=edge_y_spring,
    mode='lines',
    line=dict(width=2, color='#888'),
    hoverinfo='none',
    showlegend=False
)

# Prepare node trace
node_x_spring = []
node_y_spring = []
node_text_spring = []
node_color_spring = []

for node in tree.nodes():
    x, y = pos_spring[node]
    node_x_spring.append(x)
    node_y_spring.append(y)
    node_text_spring.append(node)
    node_color_spring.append(depths[node])

# Create node trace
node_trace_spring = go.Scatter(
    x=node_x_spring, y=node_y_spring,
    mode='markers+text',
    text=node_text_spring,
    textposition='top center',
    hoverinfo='text',
    hovertext=node_text_spring,
    marker=dict(
        size=20,
        color=node_color_spring,
        colorscale='Plasma',
        showscale=True,
        colorbar=dict(title="Depth Level"),
        line_width=2
    )
)

# Create figure
fig_tree_spring = go.Figure(data=[edge_trace_spring, node_trace_spring])

fig_tree_spring.update_layout(
    title='<b>Tree Visualization: Spring (Force-Directed) Layout</b><br><sub>Product Categories Network</sub>',
    showlegend=False,
    hovermode='closest',
    margin=dict(b=20, l=5, r=5, t=40),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    height=700,
    width=1000,
    plot_bgcolor='rgba(240, 240, 240, 0.5)'
)

fig_tree_spring.show()
print("✓ Tree Spring Layout visualization created!")
```

## 5. Graph Visualization

### 5.1 Graph Visualization Layout 1: Force-Directed (Spring) Layout

```python
# Graph Force-Directed Layout Visualization
print("Creating Graph Visualization 1: Force-Directed Layout...")

# Get positions
pos_force = layout_positions['graph']['spring']

# Prepare edge traces
edge_x_force = []
edge_y_force = []

for edge in flight_graph.edges():
    x0, y0 = pos_force[edge[0]]
    x1, y1 = pos_force[edge[1]]
    edge_x_force.extend([x0, x1, None])
    edge_y_force.extend([y0, y1, None])

# Create edge trace
edge_trace_force = go.Scatter(
    x=edge_x_force, y=edge_y_force,
    mode='lines',
    line=dict(width=1.5, color='rgba(100, 100, 100, 0.5)'),
    hoverinfo='none',
    showlegend=False
)

# Prepare node trace
node_x_force = []
node_y_force = []
node_text_force = []
node_size_force = []

# Calculate node sizes based on degree (hub airports are bigger)
degrees = dict(flight_graph.degree())
max_degree = max(degrees.values())

for node in flight_graph.nodes():
    x, y = pos_force[node]
    node_x_force.append(x)
    node_y_force.append(y)
    node_text_force.append(f"{node}<br>Routes: {degrees[node]}")
    node_size_force.append(10 + (degrees[node] / max_degree) * 25)

# Create node trace
node_trace_force = go.Scatter(
    x=node_x_force, y=node_y_force,
    mode='markers+text',
    text=[node.split('<br>')[0] for node in node_text_force],
    textposition='middle center',
    hoverinfo='text',
    hovertext=node_text_force,
    marker=dict(
        size=node_size_force,
        color='#FF6B6B',
        showscale=False,
        line_width=2,
        line_color='white'
    )
)

# Create figure
fig_graph_force = go.Figure(data=[edge_trace_force, node_trace_force])

fig_graph_force.update_layout(
    title='<b>Graph Visualization: Force-Directed Layout</b><br><sub>Flight Routes Network - Major Hubs</sub>',
    showlegend=False,
    hovermode='closest',
    margin=dict(b=20, l=5, r=5, t=40),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    height=700,
    width=1000,
    plot_bgcolor='rgba(240, 240, 240, 0.5)'
)

fig_graph_force.show()
print("✓ Graph Force-Directed Layout visualization created!")
```

### 5.2 Graph Visualization Layout 2: Circular Layout

```python
# Graph Circular Layout Visualization
print("Creating Graph Visualization 2: Circular Layout...")

# Get positions
pos_circular = layout_positions['graph']['circular']

# Prepare edge traces
edge_x_circ = []
edge_y_circ = []

for edge in flight_graph.edges():
    x0, y0 = pos_circular[edge[0]]
    x1, y1 = pos_circular[edge[1]]
    edge_x_circ.extend([x0, x1, None])
    edge_y_circ.extend([y0, y1, None])

# Create edge trace
edge_trace_circ = go.Scatter(
    x=edge_x_circ, y=edge_y_circ,
    mode='lines',
    line=dict(width=1.5, color='rgba(100, 100, 100, 0.5)'),
    hoverinfo='none',
    showlegend=False
)

# Prepare node trace
node_x_circ = []
node_y_circ = []
node_text_circ = []
node_color_circ = []

# Use betweenness centrality for coloring
betweenness = nx.betweenness_centrality(flight_graph)

for node in flight_graph.nodes():
    x, y = pos_circular[node]
    node_x_circ.append(x)
    node_y_circ.append(y)
    node_text_circ.append(f"{node}<br>Importance: {betweenness[node]:.3f}")
    node_color_circ.append(betweenness[node])

# Create node trace
node_trace_circ = go.Scatter(
    x=node_x_circ, y=node_y_circ,
    mode='markers+text',
    text=[node.split('<br>')[0] for node in node_text_circ],
    textposition='middle center',
    hoverinfo='text',
    hovertext=node_text_circ,
    marker=dict(
        size=18,
        color=node_color_circ,
        colorscale='YlOrRd',
        showscale=True,
        colorbar=dict(title="Betweenness<br>Centrality"),
        line_width=2,
        line_color='white'
    )
)

# Create figure
fig_graph_circ = go.Figure(data=[edge_trace_circ, node_trace_circ])

fig_graph_circ.update_layout(
    title='<b>Graph Visualization: Circular Layout</b><br><sub>Flight Routes Network - Airport Importance</sub>',
    showlegend=False,
    hovermode='closest',
    margin=dict(b=20, l=5, r=5, t=40),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    height=700,
    width=1000,
    plot_bgcolor='rgba(240, 240, 240, 0.5)'
)

fig_graph_circ.show()
print("✓ Graph Circular Layout visualization created!")
```
